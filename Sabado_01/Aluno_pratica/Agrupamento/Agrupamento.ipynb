```python
# Segmentação de cliente
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Dados fictícios
data = {
    "gastos_mensais": [200, 800, 150, 700, 300, 1000, 250, 900],
    "frequencia_compras": [5, 20, 3, 18, 6, 25, 4, 22]
}
df = pd.DataFrame(data)

# Modelo
kmeans = KMeans(n_clusters=2, random_state=42)
df["cluster"] = kmeans.fit_predict(df)

print(df)

# Visualização
plt.scatter(df["gastos_mensais"], df["frequencia_compras"], c=df["cluster"], cmap="coolwarm")
plt.xlabel("Gastos mensais")
plt.ylabel("Frequência de compras")
plt.show()
```

```python
import joblib

# Salvar o modelo treinado em um arquivo .pkl
joblib.dump(kmeans, 'segmentacao_clientes.pkl')
print("Modelo salvo com sucesso em 'segmentacao_clientes.pkl'")
```
